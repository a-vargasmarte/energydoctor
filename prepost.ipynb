```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
postAmps = pd.read_csv('data.csv')
preAmps = pd.read_csv('data-antes.csv')
```

```python
postAmps = postAmps.sort_index()
preAmps = preAmps.sort_index()
# temp = temp.sort_index()
# temp.head()


postAmps.head()
```

```python
def formatDates (row):
    
    if '/2019' in row['time']:
        return row['time'].replace('/2019', '/19')
        
    else:
        return row['time']
         
preAmps['time'] = preAmps.apply(lambda row: formatDates(row), axis=1)

```

```python
preAmps
```

```python
def getHour (row):
    testTime = datetime.datetime.strptime(row['time'], '%m/%d/%y %H:%M %p')
    return testTime.strftime('%H')

preAmps['hour']=preAmps.apply(lambda row: getHour(row), axis=1)

preAmps
```

```python
preAmpsByHour = preAmps.groupby(['hour']).mean()['amps']

preAmpsByHour

# preAmps
```

```python
def getWeekday (row):
    testTime = datetime.datetime.strptime(row['time'], '%m/%d/%y %H:%M %p')
    return testTime.strftime('%A')     
preAmps['weekday'] = preAmps.apply(lambda row: getWeekday(row), axis=1)
```

```python
def getPostWeekday (row):
    testTime = datetime.datetime.strptime(row['time'], '%m/%d/%y %H:%M:%S %p')
    return testTime.strftime('%A')
postAmps['weekday'] = postAmps.apply(lambda row: getPostWeekday(row), axis=1)
```

```python
def getTime (row):
    testTime = datetime.datetime.strptime(row['time'], '%m/%d/%y %H:%M %p')
    return testTime.strftime('%H:%M')     
preAmps['time'] = preAmps.apply(lambda row: getTime(row), axis=1)
```

```python
def getPostHour (row):
    testTime = datetime.datetime.strptime(row['time'], '%m/%d/%y %I:%M:%S %p')
    return testTime.strftime('%H')

postAmps['hour']=postAmps.apply(lambda row: getPostHour(row), axis=1)

postAmps
```

```python
postAmpsByHour = postAmps.groupby(['hour']).mean()['amps']

postAmpsByHour

# preAmps
```

```python
def getPostTime (row):
    testTime = datetime.datetime.strptime(row['time'], '%m/%d/%y %I:%M:%S %p')
    return testTime.strftime('%H:%M')     
postAmps['time'] = postAmps.apply(lambda row: getPostTime(row), axis=1)
```

```python
def onOrOff (row):
    if row['amps'] < 1:
        return 0
    else:
        return 1

postAmps['on'] = postAmps.apply(lambda row: onOrOff(row), axis=1)
preAmps['on'] = postAmps.apply(lambda row: onOrOff(row), axis=1)

postAmps.head()
```
